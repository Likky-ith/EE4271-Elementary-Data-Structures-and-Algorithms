```
def find_ten_largest_values(input_arr):
    output_arr=[]
    for i in range(10):
        flag1 = 0 
        for k in range(len(output_arr)):
            if (input_arr[i] > output_arr[k]) and flag1 == 0:
                output_arr.insert(k, input_arr[i])
                flag1 = 1
        if flag1 == 0:
            output_arr.append(input_arr[i])

    for i in range(10, len(input_arr)):
        flag2 = 0
        for k in range(10):
            if (input_arr[i] > output_arr[k]) and flag2 == 0:
                output_arr.insert(k, input_arr[i])
                del output_arr[-1]
                flag2 = 1

    return output_arr            
```

## *Use the divide and conquer integer multiplication algorithm to multiply the two binary integers 10011011 and 10111010*

```
def make_equal_length(str1, str2):
    len1 = len(str1)
    len2 = len(str2)
    if len1 > len2:
        for i in range(len1 -len2):
            str2= '0' + str2
    elif len2 > len1:
        for i in range(len2 - len1):
            str1= '0' + str1
    return len1
```

```
def add_bit_strings(str1, str2):
    length = make_equal_length(str1, str2)
    carry = 0
    result =''
    for i in range(length-1, -1, -1):
        bit1 = int(str1[i])
        bit2 = int(str2[i])
        sum = str((bit1 ^ bit2 ^ carry))
        result = sum + result
        carry = (bit1 & bit2)|(bit1 & carry)|(bit2 & carry)
    if (carry):
        result = '1' + result
    print("##add##", end=' ')
    print(str1, str2, carry, result)
    return result
```

## *You are given a unimodal array of ​n​ distinct elements, meaning that its entries are in increasing order up until its maximum elements, after which its elements are in decreasing order. Give an algorithm to compute the maximum element of a unimodalarray that runs in $O(logn)$  time.*

```
def find_max_value(arr, left, right):
    if right>=left:
        mid= int((left + right)/2)
        if (arr[mid] > arr[mid +1]) and (arr[mid] > arr[mid -1]):
            return arr[mid]
        elif (arr[mid] > arr[mid +1]) and (arr[mid] < arr[mid -1]):
            right = mid -1
            return find_max_value(arr, left, right)
        elif (arr[mid] < arr[mid +1]) and (arr[mid] > arr[mid -1]):
            left = mid +1
            return find_max_value(arr, left, right)
    else:
        return -1
```

```
input_array = list(map(int, input("Enter any unimodal array with distinct elements: ").split()))
left_index = 0
right_index = len(input_array) -1

max_value = find_max_value(input_array, left_index, right_index)
print(max_value)
```

## *You are given an array of n elements, and you notice that some of the elements are duplicates; that is, they appear more than once in array. Show how to remove all duplicates from the array in time $O(nlogn)$*

```
def merge(arr, l, m, r, index): 
    n1 = m - l + 1
    n2 = r- m 

    L = [0] * (n1) 
    R = [0] * (n2) 

    for i in range(0 , n1): 
        L[i] = arr[l + i]
  
    for j in range(0 , n2): 
        R[j] = arr[m + 1 + j] 
  
    # Merge the temp arrays back into arr[l..r] 
    i = 0     # Initial index of first subarray 
    j = 0     # Initial index of second subarray 
    k = l     # Initial index of merged subarray 
  
    while i < n1 and j < n2 : 
        if L[i][index] <= R[j][index]: 
            arr[k] = L[i] 
            i += 1
        else: 
            arr[k] = R[j] 
            j += 1
        k += 1

    while i < n1: 
        arr[k] = L[i] 
        i += 1
        k += 1
  
    while j < n2: 
        arr[k] = R[j] 
        j += 1
        k += 1

def mergeSort(arr,l,r, index): 
    if l < r: 
  
        # Same as (l+r)//2, but avoids overflow for 
        # large l and h 
        m = (l+(r-1))//2
  
        mergeSort(arr, l, m, index) 
        mergeSort(arr, m+1, r, index) 
        merge(arr, l, m, r, index) 
```
